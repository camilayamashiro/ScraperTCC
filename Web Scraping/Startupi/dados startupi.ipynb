```python
# Importando bibliotecas
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas

# Declarando variáveis
artigos = []

# Obtendo o HTML
req = Request('https://startupi.com.br/page/1/?s=transforma%C3%A7%C3%A3o+digital', headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req).read()
soup = BeautifulSoup(html,'html.parser')
pages = 10

# Iterando por todas as páginas do site
for i in range(pages):
    ## Obtendo o HTML
    req = Request('https://startupi.com.br/page/' + str (i + 1) + '/?s=transforma%C3%A7%C3%A3o+digital', headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html,'html.parser')
    # Obtendo as TAGs de interesse
    noticias = soup.find('section',{'class':'posts-section'}).findAll('article', {'class':'post-item'})

    # Coletando as informações das NOTICIAS
    for new in noticias:
        artigo = {}
        # Titulo
        artigo['titulo'] = new.find('h3', {'class': 'post-title'}).getText().strip()
        # Tags
        artigo['tags'] = new.find('div',{'class': 'post-meta'}).getText().strip()
        # Data
        artigo['ano']= new.find('a').img.get('src').split('/')[5]
        artigo['mes']= new.find('a').img.get('src').split('/')[6]

        #Adicionando resultado a lista de artigos
        artigos.append(artigo)
    
# Criando um dataframe com os resultados
startupi = pandas.DataFrame(artigos)
startupi.to_csv('./startupi.csv', sep=';', index = False)
startupi
```
