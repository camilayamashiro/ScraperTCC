```python
# Importando bibliotecas
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas

# Declarando variáveis
artigos = []

# Obtendo o HTML
req = Request('https://valor.globo.com/busca/?q=startup&order=recent&species=not%C3%ADcias&page=1', headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req).read()
soup = BeautifulSoup(html,'html.parser')
pages = 30

# Iterando por todas as páginas do site
for i in range(pages):
    ## Obtendo o HTML
    req = Request('https://valor.globo.com/busca/?q=startup&order=recent&species=not%C3%ADcias&page=' + str (i + 1), headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html,'html.parser')
    # Obtendo as TAGs de interesse
    noticias = soup.find('ul',{'class':'results__list'}).findAll('li', {'class':'widget widget--card widget--info'})
    
    # Coletando as informações das NOTICIAS
    for noticia in noticias:
        artigo = {}
        # Titulo
        artigo['titulo'] = noticia.find('div', {'class': 'widget--info__title'}).getText().strip()
        # Descrição
        artigo['description'] = noticia.find('p', {'class': 'widget--info__description'}).getText().strip()
        # Data
        artigo['data'] = noticia.find('div', {'class': 'widget--info__meta'}).getText().strip()

        # Adicionando resultado a lista de artigos
        artigos.append(artigo)
    
# Criando um dataframe com os resultados
valor = pandas.DataFrame(artigos)
valor.to_csv('./valor.csv', sep=';')
valor
    
    
    
```
