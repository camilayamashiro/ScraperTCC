```python
# Importando bibliotecas
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas

# Declarando variáveis
artigos = []

# Obtendo o HTML
req = Request('https://forbes.com.br/page/1/?s=transforma%C3%A7%C3%A3o+digital', headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req).read()
soup = BeautifulSoup(html,'html.parser')
pages = 10

# Iterando por todas as páginas do site
for i in range(pages):
    ## Obtendo o HTML
    req = Request('https://forbes.com.br/page/' + str (i + 1) + '/?s=transforma%C3%A7%C3%A3o+digital', headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html,'html.parser')
    # Obtendo as TAGs de interesse
    noticias = soup.find('div',{'id':'main-content'},{'class':'mh-loop mh-content'}).findAll('article', {'class':'mh-posts-list-item'})

    # Coletando as informações das NOTICIAS
    for noticia in noticias:
        artigo = {}
        # Titulo
        artigo['titulo'] = noticia.find('h3', {'class':'entry-title mh-posts-list-title'}).getText().strip()
        # Descrição
        artigo['description'] = noticia.find('div', {'class': 'mh-excerpt'}).getText().strip()
        # Data
        artigo['data'] = noticia.find('span',{'class': 'entry-meta-date updated'}).getText().strip()

        #Adicionando resultado a lista de artigos
        artigos.append(artigo)
    
# Criando um dataframe com os resultados
forbes = pandas.DataFrame(artigos)
forbes.to_csv('./forbes.csv', sep=';', index = False)
forbes
```
