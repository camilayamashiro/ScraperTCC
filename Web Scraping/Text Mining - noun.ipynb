```python
# IMPORTANDO BIBLIOTECAS
import spacy
from collections import Counter
from string import punctuation
import pt_core_news_lg
from matplotlib import pyplot as plt
import pandas as pd

# IMPORTANDO O ARQUIVO COM DADOS LIMPOS
arq = open('.\cleaned_data_vf.txt', encoding="utf8")
doc = arq.read()

# PROCESSANDO O TEXTO E CRIANDO FILTROS
nlp = pt_core_news_lg.load()
all_stopwords = nlp.Defaults.stop_words
all_stopwords |= {"transformação","digital","empresa","digitais","digitalização","brasil","empresas"}

# CRIANDO FUNÇÃO PARA OBTER PRINCIPAIS TERMOS
def get_tags(text):
    result = []
    pos_tag = ['NOUN']
    doc = nlp(text.lower())
    for token in doc:        
        if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue        
        if(token.pos_ in pos_tag):
            result.append(token.text)                
    return result

output = get_tags(doc)

# CONTANDO TERMOS MAIS CITADOS
word_counter = Counter(output)

# CRIANDO DATAFRAME COM OS RESULTADOS
df = pd.DataFrame.from_dict(word_counter, orient='index').reset_index()
df = df.rename(columns={'index':'word', 0:'count'}).sort_values(['count'], ascending=False)
df.to_csv('result_analysis_noun.csv', sep=';')

df
```
