```python
# Importando bibliotecas
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas

# Declarando variáveis
artigos = []

# Obtendo o HTML
req = Request('https://g1.globo.com/busca/?q=transforma%C3%A7%C3%A3o+digital&page=1&ajax=1', headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req).read()
soup = BeautifulSoup(html,'html.parser')
pages = 20

# Iterando por todas as páginas do site
for i in range(pages):
    ## Obtendo o HTML
    req = Request('https://g1.globo.com/busca/?q=transforma%C3%A7%C3%A3o+digital&page=' + str (i + 1) + '&ajax=1', headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html,'html.parser')
    # Obtendo as TAGs de interesse
    noticias = soup.find('div',{'class':'results__content'}).findAll('li', {'class':'widget widget--card widget--info'})
    
    # Coletando as informações das NOTICIAS
    for noticia in noticias:
        artigo = {}
        # Titulo
        artigo['titulo'] = noticia.find('div', {'class': 'widget--info__title'}).getText().strip()
        # Descrição
        artigo['description'] = noticia.find('p', {'class': 'widget--info__description'}).getText().strip()
        # Data
        artigo['data'] = noticia.find('div', {'class': 'widget--info__meta'}).getText().strip()

        #Adicionando resultado a lista de artigos
        artigos.append(artigo)
    
# Criando um dataframe com os resultados
g1 = pandas.DataFrame(artigos)
g1.to_csv('./g1.csv', sep=';')
g1
    
    
    
```
