```python
# Importando bibliotecas
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas

# Declarando variáveis
artigos = []

# Obtendo o HTML
req = Request('https://www.moneytimes.com.br/page/1/?s=transforma%C3%A7%C3%A3o+digital', headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req).read()
soup = BeautifulSoup(html,'html.parser')
pages = 20

# Iterando por todas as páginas do site
for i in range(pages):
    ## Obtendo o HTML
    req = Request('https://www.moneytimes.com.br/page/' + str (i + 1) + '/?s=transforma%C3%A7%C3%A3o+digital', headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html,'html.parser')
    # Obtendo as TAGs de interesse
    noticias = soup.find('main',{'class':'main'}).findAll('div', {'class':'news-item'})

    # Coletando as informações das NOTICIAS
    for new in noticias:
        artigo = {}
        # Titulo
        artigo['titulo'] = new.find('h2', {'class': 'news-item__title'}).getText().strip()
        # Data
        artigo['data'] = new.find('span',{'class': 'date'}).getText().strip()

        #Adicionando resultado a lista de artigos
        artigos.append(artigo)
    
# Criando um dataframe com os resultados
money = pandas.DataFrame(artigos)
money.to_csv('./money.csv', sep=';')
money
```
